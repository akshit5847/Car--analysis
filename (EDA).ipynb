```python
'''
Quantiles are the set of values/points that divides the dataset into groups of equal size.
For example, in the figure, there are nine values that splits the dataset. 
Those nine values are quantiles.
'''
# Select only numeric columns
numeric_columns = duplicate_rows.select_dtypes(include='number')  # we use this because  we want only numeric values.
Q1 = numeric_columns.quantile(0.25)   # it shows 25 percentile of the duplicate_rows 
Q3 = numeric_columns.quantile(0.75)   # it shows 75 percentile of the duplicate_rows
IQR = Q3 - Q1                         # it shows 75-25 percentile  of the duplicate_rows
IQR   
```

```python
# lower limit
lower_limit = Q1 - 1.5 * IQR
# upper limit
upper_limit = Q3 + 1.5 * IQR
lower_limit , upper_limit
```

TypeError: unsupported operand type(s) for -: 'str' and 'str': This error occurs if your DataFrame contains non-numeric columns. quantile() operates on numeric data, so if your DataFrame includes non-numeric columns, you should either remove them or select only the numeric columns for your analysis.

```python

df_1=numeric_columns[~((numeric_columns < (lower_limit)) | (numeric_columns > (upper_limit)))]

# or we can directly do that too like this
df_2= numeric_columns[~((numeric_columns < (Q1 - 1.5 * IQR)) | (numeric_columns > (Q3 + 1.5 * IQR))).any(axis=1)] 
                    # it gives an outlier values of our data , when we use ' | ' operator  it gives an outliers of our data

df_1.shape , df_2.shape # we got different value because we use ".any(axis=1)"

```

```python
df_1
```

```python
df_3=numeric_columns[~((numeric_columns < (lower_limit)) & (numeric_columns > (upper_limit)))]# here we get the value of our
                                                                     #  data without outliers , when we use '&' operator 
                                                                    # it gives the value without outliers               
df_3
```
